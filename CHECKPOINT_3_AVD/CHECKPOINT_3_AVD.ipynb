```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/drive/MyDrive/AVD/mobile_dataset_palingbaru.xlsx")
df.head()
```

# Menganalisis Trean Rata-Rata Harga Ponsel Berdasarkan Tahun Rilisnya

```python
#menampilkan line-chart
plt.figure(figsize=(8, 5))
df["Launched Price in USA (USD)"] = pd.to_numeric(df["Launched Price in USA (USD)"], errors='coerce')
avg_price_per_year = df.groupby(df["Launched Year"])["Launched Price in USA (USD)"].mean()

plt.plot(avg_price_per_year.index, avg_price_per_year.values, marker="o", linestyle="-", color="purple")
plt.xlabel("Tahun Rilis")
plt.ylabel("Rata-rata Harga (USD)")
plt.title("Tren Rata-rata Harga Ponsel Berdasarkan Tahun Rilis")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

## Insight Yang Diambil Berdasarkan Grafik:

1. Kenaikan Harga Rata-rata:

Jika grafik menunjukkan tren naik, berarti harga rata-rata ponsel semakin mahal setiap tahun.
Kemungkinan penyebabnya: teknologi semakin canggih (chipset lebih cepat, kamera lebih baik, layar lebih maju).

2. Penurunan di Tahun Tertentu:

Jika ada penurunan harga di tahun tertentu, mungkin karena:
Meningkatnya persaingan (brand seperti Xiaomi & Realme menekan harga).
Inovasi baru yang membuat ponsel lebih murah untuk diproduksi.
Faktor ekonomi global yang mempengaruhi daya beli konsumen.

3. Stagnasi atau Fluktuasi:

Jika harga tidak banyak berubah atau naik-turun, mungkin ada faktor lain seperti:
Brand lebih fokus ke kelas menengah, bukan flagship mahal.
Siklus inovasi yang melambat (misalnya tidak ada terobosan besar seperti 5G dulu).

## Tindakan Yang Dapat Diambil Berdasarkan Grafik:

1. Jika harga terus naik, lebih baik membeli ponsel sebelum model terbaru dirilis, biasanya akan ada diskon dan harga model lama akan turun.

2. Perhatikan tren harga untuk memilih waktu terbaik membeli (misalnya saat harga cenderung turun).

# Mengkategorikan Ponsel Menjadi Top 10 Berdasarkan Spesifikasi Terbaik

```python
#kategorikan ke ponsel
df_phones = df[df["Screen Size (Inches)"] <= 8]
top_phones = df_phones.loc[
    (df_phones["RAM (GB)"] * 0.4 + df_phones["Battery Capacity (mAh)"] * 0.3 + df_phones["Screen Size (Inches)"] * 0.3).nlargest(10).index
]

#menampilkan bar-chart
plt.figure(figsize=(10, 6))
spek_score = top_phones["RAM (GB)"] * 0.4 + top_phones["Battery Capacity (mAh)"] * 0.3 + top_phones["Screen Size (Inches)"] * 0.3

ax = sns.barplot(
    x=spek_score,
    y=top_phones["Model Name"],
    palette="viridis"
)

#menampilkan harga
for index, value in enumerate(top_phones["Launched Price in USA (USD)"]):
    ax.text(
        (top_phones["RAM (GB)"].iloc[index] * 0.4 + top_phones["Battery Capacity (mAh)"].iloc[index] * 0.3 + top_phones["Screen Size (Inches)"].iloc[index] * 0.3) + 1,
        index, f"${int(value)}", va='center', fontsize=10, color='black'
    )
#judul tabel
print(top_phones)
plt.xlabel("Skor Spesifikasi")
plt.ylabel("Nama Ponsel")
plt.title("10 Ponsel dengan Spesifikasi Terbaik")

plt.show()
```

## Insight Yang Dapat Diambil Berdasarkan Chart Tersebut Adalah:

1. Ponsel dengan Spesifikasi Tertinggi

Ponsel dengan skor spesifikasi tertinggi memiliki kombinasi RAM besar, baterai besar, layar yang ideal, dan prosesor yang cepat.
Jika ada merek yang mendominasi di posisi teratas, itu menunjukkan mereka lebih unggul dalam teknologi hardware.

2. Harga dan Performa

Jika harga ponsel-ponsel ini tinggi, maka spesifikasi premium memang sejalan dengan harga.
Jika ada ponsel dengan skor tinggi tetapi harganya lebih murah dibanding yang lain, itu bisa menjadi pilihan terbaik untuk value for money.

## Tindakan Yang Dapat Diambil Berdasarkan Chart Tersebut Adalah:

1. Jika konsumen memiliki budget lebih dan mencari ponsel dengan spesifikasi terbaik, maka ponsel dari chart tersebut dapat menjadi pilihan yang baik, terutama yang memiliki harga lebih kompetitif.

2. Jika konsumen ingin mencari ponsel dengan baterai yang tahan lama, maka cek apakah ponsel dengan baterai besar mendominasi peringkat teratas, dan jika lebih suka ponsel compact, cari model yang tetap memiliki skor tinggi meskipun layarnya lebih kecil.

# Mengkategorikan Ponsel Menjadi Top 10 Berdasarkan Harga dan Spesifikasi

```python
#kategorikan ke ponsel
df_phones = df[df["Screen Size (Inches)"] <= 8]

#memilih ponsel termurah
harga_batas = df_phones["Launched Price in USA (USD)"].quantile(0.3)
df_murah = df_phones[df_phones["Launched Price in USA (USD)"] <= harga_batas]
top_phones = df_murah.loc[
    (df_murah["RAM (GB)"] * 0.4 + df_murah["Battery Capacity (mAh)"] * 0.3 + df_murah["Screen Size (Inches)"] * 0.3).nlargest(10).index
]

#bar-chart
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x=(top_phones["RAM (GB)"] * 0.4 + top_phones["Battery Capacity (mAh)"] * 0.3 + top_phones["Screen Size (Inches)"] * 0.3),
    y=top_phones["Model Name"],
    palette="viridis"
)

#menampilkan harga
for index, value in enumerate(top_phones["Launched Price in USA (USD)"]):
    ax.text(
        (top_phones["RAM (GB)"].iloc[index] * 0.4 + top_phones["Battery Capacity (mAh)"].iloc[index] * 0.3 + top_phones["Screen Size (Inches)"].iloc[index] * 0.3) + 1,
        index, f"${int(value)}", va='center', fontsize=10, color='black'
    )
#judul tabel
print(top_phones)
plt.xlabel("Skor Spesifikasi")
plt.ylabel("Nama Ponsel")
plt.title("10 Ponsel dengan Spek Terbaik & Harga Terjangkau")

plt.show()


```

## Insight Yang Dapat Diambil Berdasarkan Chart Di Atas Adalah:

1. Ponsel Murah dengan Spesifikasi Tinggi

Ponsel dalam daftar ini memiliki kombinasi RAM yang cukup besar, baterai tahan lama, dan ukuran layar ideal, tetapi dengan harga yang tetap terjangkau.
Jika ada merek yang mendominasi, itu menunjukkan mereka memiliki strategi harga yang agresif untuk segmen ini.

2. Harga vs Performa

Jika ada ponsel dengan skor spesifikasi tinggi tetapi harga lebih rendah dibanding pesaingnya, berarti ponsel itu adalah opsi terbaik untuk budget rendah.
Jika selisih skor spesifikasi antara peringkat atas dan bawah cukup kecil, berarti persaingan di kelas entry-level cukup ketat.

3. Tren di Ponsel Terjangkau

Jika ponsel dengan RAM lebih besar atau baterai lebih besar mendominasi daftar, berarti fitur ini lebih diutamakan di segmen harga murah.
Jika ada ponsel dengan layar lebih besar tetapi tetap di harga murah, itu bisa menarik bagi pengguna yang suka layar luas tanpa harus membeli flagship.

## Tindakan Yang Dapat Dilakukan Berdasarkan Chart DI Atas Adalah:

1. Jika konsumen mencari ponsel terbaik dan memiliki  budget yang terbatas, maka pilihlah ponsel dari daftar ini karena memiliki spesifikasi terbaik di kelasnya dengan harga yang cukup murah.

2. Jika ingin ponsel dengan daya tahan baterai lama, cari model dengan kapasitas baterai yang lebih besar, dan jika lebih suka ponsel compact, perhatikan model dengan layar lebih kecil tetapi tetap memiliki skor tinggi.

3. Meskipun memiliki harga yang cukup murah dengan spesifikasi yang bagus, konsumen juga perlu melakukan riset terlebih dahulu, seperti apakah ponsel dalam chart dapat digunakan untuk waktu yang lama? dan apakah ponsel dari merek tersebut memiliki after sales yang cukup baik agar konsumen dapat memiliki ponsel yang bagus meskipun dengan harga yang cukup murah.

# Korelasi Antar Variabel Numerik

```python
#kategorikan ke ponsel
df_phones = df[df["Screen Size (Inches)"] <= 8]
numerical_df = df_phones.select_dtypes(include=["number"])

plt.figure(figsize=(8, 6))
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

#judul tabel
plt.title("Heatmap Korelasi Antar Variabel Numerik (Hanya Ponsel)")
plt.show()

```

## Insight dan Tindakan Yang Dapat Diambil Berdasarkan Chart Tersebut Adalah:

1. Korelasi Tinggi antara RAM dan Harga

Insight :

Jika korelasi antara RAM (GB) dan Harga mendekati 1.0, berarti RAM adalah faktor utama yang menentukan harga ponsel.
Ponsel dengan RAM lebih besar umumnya lebih mahal karena menunjang multitasking dan performa yang lebih baik.

Tindakan:

Jika ingin ponsel dengan performa tinggi (gaming, editing, multitasking), pilih yang RAM lebih besar, meskipun harganya lebih tinggi.


2. Korelasi Lemah atau Negatif antara Kapasitas Baterai dan Harga

Insight:

Jika korelasi antara Battery Capacity (mAh) dan Harga rendah atau negatif, berarti kapasitas baterai besar tidak selalu berbanding lurus dengan harga.
Ini menunjukkan bahwa ponsel dengan baterai besar bisa ditemukan di berbagai segmen harga, termasuk ponsel murah.

Tindakan:
Kita bisa menemukan ponsel dengan baterai besar tanpa perlu membayar mahal

3. Korelasi Positif antara Ukuran Layar dan Harga

Insight:

Jika korelasi Layar dan Harga cukup tinggi, berarti ponsel dengan layar lebih besar cenderung lebih mahal.
Hal ini bisa terjadi karena layar besar sering dikombinasikan dengan spesifikasi premium seperti refresh rate tinggi atau panel OLED.

Tindakan:

Jika ingin memiliki ponsel dengan harga yang lebih murah, mungkin bisa memilih layar yang lebih kecil dengan spesifikasi serupa.

4. Korelasi Tinggi antara Prosesor dengan RAM & Harga

Insight:

Jika prosesor memiliki korelasi tinggi dengan RAM dan Harga, berarti ponsel dengan prosesor lebih canggih cenderung memiliki RAM lebih besar dan lebih mahal.
Ini menunjukkan bahwa prosesor high-end biasanya hanya ditemukan di ponsel flagship atau premium.

Tindakan:

Jika mencari ponsel dengan performa tinggi, pastikan memilih chipset terbaru, meskipun harus membayar lebih mahal.
